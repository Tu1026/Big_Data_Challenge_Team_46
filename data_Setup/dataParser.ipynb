```python
import pandas as pd
import os
from pathlib import Path
import re
import numpy as np
import sqlite3
from contextlib import closing
```

```python
count = 0
with closing(sqlite3.connect('BigDataChallenge.sqlite')) as cnxn:
    for root, dirs, files in os.walk(Path('.').resolve().absolute()):
        for file in files:
            if re.match(".*Summary\_Details\.csv", file):
                temp_detail = pd.read_csv(os.path.join(root,file), engine='python')
                if count == 0:
                    temp_detail.to_sql('Details', cnxn, method='multi', if_exists='replace', index=False, chunksize=30000)
                    cnxn.commit()
                    count += 1
                else:
                    print(file)
                    temp_detail.to_sql('Details', cnxn, method='multi', if_exists='append', index=False, chunksize=30000)
                    cnxn.commit()
```

```python
count = 0
for root, dirs, files in os.walk(Path('.').resolve().absolute()):
    for file in files:
        if re.match(".*Summary\_Hashtag\.csv", file):
            temp_hash = pd.read_csv(os.path.join(root,file), engine='python')
            with closing(sqlite3.connect('BigDataChallenge.sqlite')) as cnxn:
                if count == 0:
                    temp_hash.to_sql('Hashtag',  cnxn, method='multi', if_exists='replace', index=False, chunksize=10000)
                    cnxn.commit()
                    count += 1
                else:
                    print(file)
                    temp_hash.to_sql('Hashtag',  cnxn, method='multi', if_exists='append', index=False, chunksize=10000)
                    cnxn.commit()
```

```python
count = 0
with closing(sqlite3.connect('BigDataChallenge.sqlite')) as cnxn:
    for root, dirs, files in os.walk(Path('.').resolve().absolute()):
        for file in files:
            if re.match("202(1\_0[1-4]|(0\_0[8-9]|0\_1[0-2])).*Summary\_Sentiment\.csv", file):
                temp_sent = pd.read_csv(os.path.join(root,file), engine='python')
                if count == 0:
                    temp_sent.to_sql('Sentiment', cnxn, method='multi', if_exists='replace', index=False, chunksize=10000)
                    count += 1
                else:
                    print(file)
                    temp_sent.to_sql('Sentiment', cnxn, method='multi', if_exists='append', index=False, chunksize=10000)
        cnxn.commit()
```
